```python
def airline_seat_reservation():


    num_tickets = int(input("\n How Many Tickets You Want: "))
    seats = [0] * 10  # Initialize all seats as unoccupied (0)

    for _ in range(num_tickets):
        
        print("   Class\n\n   [1] Business Class\n   [2] Economic class\n\n")
        
        class_choice = int(input("   Choose: "))

        if class_choice == 1:
            
            print("   Business class\n\n   Seats available\n     |1|2|3|4|5|\n\n")
            
            seat_available = False

            while not seat_available:
                name = input("   Enter Passenger Name: ")
                passport_number = int(input("\n Enter Your PassPortNumber : "))
                place = input("\n   Enter Your Place :")
                destination = input("\n Enter Your Destination :")
                seat_choice = int(input("   Pick a seat: "))

                if seat_choice < 1 or seat_choice > 5:
                    print("\n\n   Invalid seat number. Please choose a seat between 1 and 5.")
                    continue

                if seats[seat_choice - 1] == 0:  # Check if seat is available
                    seats[seat_choice - 1] = 1  # Mark seat as occupied
                    seat_available = True
                else:
                    print("\n\n   Seat taken.\n\n")

           
            print("\n")
            print(" Class: Business class")
            print(" Passenger Name : ", name)
            print("  PassPort ID :", passport_number)
            print("   ", place, "----- > ", destination)
            print(" Seat no : ", seat_choice, "\n")

        elif class_choice == 2:
            
            print("   Economical class\n\n   Seats available\n |6|7|8|9|10|\n\n")
            
            seat_available = False

            while not seat_available:
                name = input("   Enter Passenger Name: ")
                passport_number = int(input("\n    Enter Your PassPortNumber : "))
                place = input("\n Enter Your Place :")
                destination = input("\n   Enter Your Destination :")
                seat_choice = int(input("   Pick a seat: "))

                if seat_choice < 6 or seat_choice > 10:
                    print("\n\n   Invalid seat number. Please choose a seat between 6 and 10.")
                    continue

                if seats[seat_choice - 1] == 0:  # Check if seat is available
                    seats[seat_choice - 1] = 1  # Mark seat as occupied
                    seat_available = True
                else:
                    print("\n\n   Seat taken.\n\n")

            
            print("\n")
            print(" Class: Economical class")
            print("  Passenger Name : ", name)
            print("   PassPort ID :", passport_number)
            print("   ", place, " > ", destination)
            print("   Seat no : ", seat_choice, "\n")

        else:
            print("\nWrong number!No seat for you!\n\n")

if __name__ == "__main__":
    airline_seat_reservation()
```
